## 1.2. Calcular la corriente el√©ctrica en el conductor circular.

Primero se identifica que se tiene un claro caso de la ley de inducci√≥n de faraday ya que tenemos una densidad de flujo magn√©tico que genera un voltaje en una espira.

Entonces para comenzar se necesita hallar ese voltaje que es la FEM inducida en la espira el cual se calcula mediante el flujo magnetico sobre la suberficie de la espira con la siguiente ecuaci√≥n: 

$$ e = -N \iint_S \frac{\partial \vec{B} }{ \partial t} ds$$

Se halla el flujo magn√©tico $\frac{\partial \vec{B} }{ \partial t}$ de la siguiente manera:


    
$$ \frac{\partial \vec{B} }{ \partial t} = \frac{\partial }{ \partial t} ( 0.5\sin10^3t(-z)) = 500\cos(10^3t)(-\hat{z}) $$

$$ \frac{\partial \vec{B} }{ \partial t} = 500\cos(10^3t)(-\hat{z}) $$
    
Ya hallado el flujo magnetico ahora se van a cuantizar las lineas del flujo magn√©tico que estan pasando por la superficie circular del conductor como se muestra a continuaci√≥n, donde se escoge la direcci√≥n del $ds$ en $(-\hat{z})$, la cual es la misma direcci√≥n de $\vec{B}$: 

$$ e = -N \iint_S \frac{\partial \vec{B} }{ \partial t} ds$$
    
  $$ e = -N \iint_S 500\cos(10^3t)(-\hat{z}) ds(-\hat{z})$$

Multiplicando los vectores unitarios: 

$$(-\hat{z})(-\hat{z})=1$$

Se tiene qu√©:

  $$ e = -N \iint_S 500\cos(10^3t) ds$$

Se sabe que la integral no depende del tiempo entonces se sacan todos lo que no dependa de ds:

   $$ e = -N \times 500\cos(10^3t)) \iint_S  ds$$

Se sabe que se quiere hallar el flujo en la superficie, por lo tanto la integral da como resultado el √°rea del circulo que depende del radio que el problema nos ilustra

  $$ e = -N \times 500\cos(10^3t) \times A_\circ $$
  $$ e = -N \times 500\cos(10^3t) \times \pi r^2 $$
    
   Se tiene que $N=1$ y $r = 1.5m$, entonces:
   
  $$ e = -1 \times 500\cos(10^3t) \times \pi \times 1.5^2 $$
   $$ e = -1125 \pi \cos(10^3t)[v] $$
    
Ahora para hallar la corriente en el conductor se utiliza la ley de ohm:

$$ V=RI $$

$$ I=\frac{V}{R} $$

Entonces con $ V= e = -1125 \pi \cos(10^3t)[v] $ y $R=0.0 5 ùêæ[\Omega]$

$$ I=\frac{-1125 \pi \cos(10^3t)[v]}{0. 05 ùêæ[\Omega]} $$

$$ I=- \frac{45}{2} \pi \cos(10^3t)[A] $$

$$ I=- 22.5 \pi \cos(10^3t)[A] $$

## 2.1. Determinar las componentes de la intensidad de campo magn√©tico $\vec{H}$.

Se identifica que hay un campo el√©ctrico en el vac√≠o, entonces podemos hacer uso de la ley de Gauss para campo magn√©tico para determinar el campo magn√©tico de la sigiente forma:

$$ \nabla \times \vec{E} = -\frac{\partial}{\partial t} \vec{B} $$

Hallamos el rotacional del campo el√©ctrico para luego hallar el campo mgan√©tico.



$$ \nabla \times \vec{E} = det\begin{equation}
\begin{bmatrix}
\hat{x} & \hat{y} & \hat{z}\\
\frac{\partial}{\partial x} & \frac{\partial}{\partial y} & \frac{\partial}{\partial z}\\
\vec{E_x} & \vec{E_y} & \vec{E_z}
\end{bmatrix}
 =  det
\begin{bmatrix}
\hat{x} & \hat{y} & \hat{z}\\
\frac{\partial}{\partial x} & \frac{\partial}{\partial y} & \frac{\partial}{\partial z}\\
\vec{E_x} & 0 & 0
\end{bmatrix}
\end{equation}  $$

Se sabe que el campo el√©ctrico solo tiene componente en $\hat{x}$ entonces las demas componentes son cero

$$ \nabla \times \vec{E} = det\begin{bmatrix}
\hat{x} & \hat{y} & \hat{z}\\
\frac{\partial}{\partial x} & \frac{\partial}{\partial y} & \frac{\partial}{\partial z}\\
\vec{E_x} & 0 & 0
\end{bmatrix} =(0-0)\hat{x}-(0-\frac{\partial}{\partial z}\vec{E_x})\hat{y}+(0-\frac{\partial}{\partial y}\vec{E_x})\hat{z} $$

Como $E_x$ no depende de y, entonces la $\frac{\partial}{\partial y}\vec{E_x}=0$

$$ \nabla \times \vec{E} = 0\hat{x}-(-\frac{\partial}{\partial z}\vec{E_x})\hat{y}+0\hat{z} =(\frac{\partial}{\partial z}\vec{E_x})\hat{y} $$

Entonces hay que derivar el campo el√©ctrico en $\hat{x}$ con respecto de $z$.

$$ \nabla \times \vec{E} = (\frac{\partial}{\partial z}\vec{E_x})\hat{y} = \Big[\frac{\partial}{\partial z}\Big( E_0\cos\big(\omega(t‚àí\frac{z}{v})\big)\Big)\Big]\hat{y} $$

Entonces se tiene que el rotor de $\vec{E}$ es:



$$ \nabla \times \vec{E} =  \frac{E_0\omega}{v}\sin \left(\omega\left(t-\frac{z}{v}\right)\right)\hat{y}$$

Por lo tanto:

$$\nabla \times \vec{E} =  -\frac{\partial}{\partial t} \vec{B} =  \frac{E_0\omega}{v}\sin \left(\omega\left(t-\frac{z}{v}\right)\right)\hat{y}$$

Entonces se puede hallar el campo magn√©tico $\vec{B}$ integrando el rotacional con respecto a $dt$ de la siguiente manera:

$$-\frac{\partial}{\partial t} \vec{B} =  \nabla \times \vec{E}$$

$$-\int \frac{\partial}{\partial t} \vec{B} dt = \int (\nabla \times \vec{E} )dt$$

$$\vec{B}  = - \int (\nabla \times \vec{E}) dt$$

Entonces integrando se tiene qu√©:

$$\vec{B}  = - \int \Big[\frac{E_0\omega}{v}\sin \left(\omega\left(t-\frac{z}{v}\right)\right)\hat{y}\Big] dt$$

Entonces el campo magn√©tico $\vec{B}$ es:

$$\vec{B}  =  \frac{E_0}{v}\cos \left(\omega\left(t-\frac{z}{v}\right)\right)\hat{y}$$


Ya con el campo magn√©tico hallado solo resta utilizar la ecuaci√≥n constitutiva $\vec{B}=\mu \vec{H}$ para hallar $\vec{H}$ como se muestra a continuaci√≥n:

$$\vec{H}=\frac{\vec{B}}{\mu} $$ 

Como los campos est√°n een el vac√≠o entonces:

$$\vec{H}=\frac{\vec{B}}{\mu_0} $$ 

La intensidad de campo magn√©tico es:

$$\vec{H}=\frac{E_0}{\mu_0v}\cos \left(\omega\left(t-\frac{z}{v}\right)\right)\hat{y} $$

## 2.2. Determinar la siguiente relaci√≥n $\frac{\vec{E\hat{x}}} {\vec{H\hat{y}} }$, tomando; $ùë£= \frac{1}{\sqrt{\mu_0\varepsilon_0}}$

Se procede a dividir la componente $\hat{x}$ de $\vec{E}$ sobre la componente $\hat{y}$ de $\vec{H}$ como se muestra a continuaci√≥n:

$$ \frac{\vec{E\hat{x}}} {\vec{H\hat{y}} } = \frac{E_0\cos\big(\omega(t‚àí\frac{z}{v})\big)}{\frac{E_0}{\mu_0 v}\cos \big(\omega(t-\frac{z}{v})\big)} $$

$$ \frac{\vec{E\hat{x}}} {\vec{H\hat{y}} } = \frac{\frac{E_0}{1}}{\frac{E_0}{\mu_0 \frac{1}{\sqrt{\mu_0\varepsilon_0}}}} = \frac{\mu_0}{\sqrt{\mu_0\varepsilon_0}}$$

$$ \frac{\vec{E\hat{x}}} {\vec{H\hat{y}} } =\sqrt{\frac{\mu_0}{\varepsilon_0}}$$

## 2.4. Graficar la intensidad de campo magn√©tico $\vec{H}$ , s√≠:  $E_0=0.5$; $f=180kHz$; $ùë£= \frac{1}{\sqrt{\mu_0\varepsilon_0}}$; $z=0.1$; $\mu_0=4\pi10^{‚àí7}$; $\varepsilon_0= \frac{1}{36\pi 10^{‚àí9}}$ en una ventana temporal de $100ùëöùë†$.

Se va a graficar exactamente igual que en el punto 1.3



```
import matplotlib.pyplot as plt
import numpy as np

#Primero definimos el vector de tiempo que va a generar la ventana de 0s a 100ms

t=np.linspace(0,100*10**(-3),1000)

#damos los valores de E0, f, v, z_0, mu_0, ep_0 

E=0.5
f=180*10**3
w=2*np.pi*f
mu_0=4*np.pi*10**(-7)
ep_0=1/(36*np.pi*10**(-9))
z_0=0.1
v=1/np.sqrt(mu_0*ep_0)

#ahora definimos la funci√≥n de intensidad de campo

H=(E/(mu_0*v))*np.cos(w*(t-z_0/v))

#Finalmente se grafica

plt.plot(t,H, label="$H$")
plt.legend()
plt.xlabel('$t[s]$')
plt.grid()
plt.ylabel('$H$')
plt.title("Intensidad de Campo Magn√©tico en el tiempo")
plt.show()
```
