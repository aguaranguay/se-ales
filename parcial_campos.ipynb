## 1.1. Graficar el campo vectorial $\vec{ùêµ}$ (espacial).

```
#Se inicializan las librerias a utilizar

from mpl_toolkits.mplot3d import axes3d
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure()
ax = fig.gca(projection='3d') 

#Se arma el plano x,y,z

x, y, z = np.meshgrid(np.arange(-0.8, 1, 0.2),   #Devuelve matrices de coordenadas a vectores de coordenadas.Crea matrices de coordenadas para evaluaciones vectorizadas de campos escalares / vectoriales
                         np.arange(-0.8, 1, 0.2),
                         np.arange(-0.8, 1, 0.8))
#Se dan los valores de cada componente en este caso u,v,w

u = 0 
v = 0
w = -0.5*np.sin(10**3)

#Finalmente se grafica

ax.quiver(x, y, z, u, v, w, length=0.2, color = 'green')

plt.show()
```

## 1.2. Calcular la corriente el√©ctrica en el conductor circular.

Primero se identifica que se tiene un claro caso de la ley de inducci√≥n de faraday ya que tenemos una densidad de flujo magn√©tico que genera un voltaje en una espira.

Entonces para comenzar se necesita hallar ese voltaje que es la FEM inducida en la espira el cual se calcula mediante el flujo magnetico sobre la suberficie de la espira con la siguiente ecuaci√≥n: 

$$ e = -N \iint_S \frac{\partial \vec{B} }{ \partial t} ds$$

Se halla el flujo magn√©tico $\frac{\partial \vec{B} }{ \partial t}$ de la siguiente manera:


    
$$ \frac{\partial \vec{B} }{ \partial t} = \frac{\partial }{ \partial t} ( 0.5\sin10^3t(-z)) = 500\cos(10^3t)(-\hat{z}) $$

$$ \frac{\partial \vec{B} }{ \partial t} = 500\cos(10^3t)(-\hat{z}) $$
    
Ya hallado el flujo magnetico ahora se van a cuantizar las lineas del flujo magn√©tico que estan pasando por la superficie circular del conductor como se muestra a continuaci√≥n, donde se escoge la direcci√≥n del $ds$ en $(-\hat{z})$, la cual es la misma direcci√≥n de $\vec{B}$: 

$$ e = -N \iint_S \frac{\partial \vec{B} }{ \partial t} ds$$
    
  $$ e = -N \iint_S 500\cos(10^3t)(-\hat{z}) ds(-\hat{z})$$

Multiplicando los vectores unitarios: 

$$(-\hat{z})(-\hat{z})=1$$

Se tiene qu√©:

  $$ e = -N \iint_S 500\cos(10^3t) ds$$

Se sabe que la integral no depende del tiempo entonces se sacan todos lo que no dependa de ds:

   $$ e = -N \times 500\cos(10^3t)) \iint_S  ds$$

Se sabe que se quiere hallar el flujo en la superficie, por lo tanto la integral da como resultado el √°rea del circulo que depende del radio que el problema nos ilustra

  $$ e = -N \times 500\cos(10^3t) \times A_\circ $$
  $$ e = -N \times 500\cos(10^3t) \times \pi r^2 $$
    
   Se tiene que $N=1$ y $r = 1.5m$, entonces:
   
  $$ e = -1 \times 500\cos(10^3t) \times \pi \times 1.5^2 $$
   $$ e = -1125 \pi \cos(10^3t)[v] $$
    
Ahora para hallar la corriente en el conductor se utiliza la ley de ohm:

$$ V=RI $$

$$ I=\frac{V}{R} $$

Entonces con $ V= e = -1125 \pi \cos(10^3t)[v] $ y $R=0.0 5 ùêæ[\Omega]$

$$ I=\frac{-1125 \pi \cos(10^3t)[v]}{0. 05 ùêæ[\Omega]} $$

$$ I=- \frac{45}{2} \pi \cos(10^3t)[A] $$

$$ I=- 22.5 \pi \cos(10^3t)[A] $$


$$ I=- 70.68583471 \cos(10^3t)[A] $$

## 1.3. Graficar la corriente el√©ctrica en una ventana temporal de 100ùëöùë†.

Se va a graficar una ventana de la corriente en el tiempo

```
import matplotlib.pyplot as plt
import numpy as np

#Primero definimos el vector de tiempo que va a generar la ventana de 0s a 100ms

t=np.linspace(0,100*10**(-3),1000)

#ahora definimos la funci√≥n de corriente

i=-22.5*np.pi*np.cos(10**3*t)

#Finalmente se grafica

plt.plot(t,i, label="$i(t)$")
plt.legend()
plt.xlabel('$t[s]$')
plt.grid()
plt.ylabel('$i(t)$')
plt.title("Corriente en el tiempo")
plt.show()
```

## 1.4. Graficar la corriente si el radio del conductor circular cambia a 3.7ùëö.

Primero Calculamos la FEM con el radio propuesto por el ejercicio

$$ e = -N \times 500\cos(10^3t) \times \pi r^2 $$

$$ e = -500\cos(10^3t) \times \pi \times (3.7)^2 $$

$$ e = -6845\pi\cos(10^3t) $$

Y ahora se calcula la corriente de la misma manera que en el punto 1.2 pero con $V=e=-6845\pi\cos(10^3t)$

$$ I=\frac{V}{R} $$

$$ I=\frac{-6845\pi\cos(10^3t)[v]}{0. 05 ùêæ[\Omega]} $$

$$ I=- 136.9 \pi \cos(10^3t)[A] $$

$$ I=- 430.0840343 \cos(10^3t)[A] $$

Ya con la nueva funci√≥n corriente se va a gr√°ficar de la misma manera que en el punto anterior

```
#No volvemos a definir el vector tiempo por que ya est√° definido, entonces solo definimos la funci√≥n de corriente

i=-136.9*np.pi*np.cos(10**3*t)

#Finalmente se grafica

plt.plot(t,i, label="$i(t)$")
plt.legend()
plt.xlabel('$t[s]$')
plt.grid()
plt.ylabel('$i(t)$')
plt.title("Corriente en el tiempo")
plt.show()
```

Se nota que la forma de la onda de corriente es la misma para ambos casos, pero la amplitud de la corriente aumenta para el nuevo radio, entonces se puede decir, que entre m√°s radio hay, m√°s voltaje, y por lo tanto m√°s corriente.

## 1.5. Calcular nuevamente la corriente el√©ctrica para el siguiente campo magn√©tico.

$$\vec{ùêµ} =3\sqrt{2}\sin 10^3ùë° (‚àí\hat{x})$$

Multiplicando los vectores unitarios: 

$$(-\hat{x})(-\hat{z})=0$$

  $$ e = 0 $$
